```python
target_map = {
    'STAR': 0,
    'GALAXY': 1,
    'QSO': 2,
}
```

```python
import pandas as pd

data = pd.read_csv('data/SDSS_DR18.csv')
data['class'] = data['class'].map(target_map)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.drop(['objid', 'specobjid', 'fiberid', 'run'], axis=1)
```

```python
x = data.drop('class', axis=1)  # Features
y = data['class']               # Target (Celestial Object Class)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='class', data=data)
plt.title('Distribution of Celestial Object Types')
plt.show()
```

```python
sns.pairplot(data.sample(1000), hue='class')
plt.show()
```

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.discriminant_analysis import StandardScaler


scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier


model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix


y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
